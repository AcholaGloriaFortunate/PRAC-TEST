```python
import matplotlib.pyplot as plt
```

```python
year = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
christian = [77, 76, 75, 73, 73, 71, 69, 68, 67, 65]
unaffiliated = [17, 17, 19, 19, 20, 21, 24, 23, 25, 26]
```

```python
plt.plot(year, christian, label='Christian')
plt.plot(year, unaffiliated, label='Unaffiliated')

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Christian and Unaffiliated Over Time in Africa')
plt.legend()

plt.show()
```

```python
#Using the matplotlib.pyplot, the x-values are the years from 2009 t0 2018 and the y values are the percentages
#christians and unaffiliated individuals in Africa.
#The blue line represents the trend in the percentage of Christians over time, 
# while the orange line represents the trend in the percentage of Unaffiliated individuals.
#From the chart, we can observe that the percentage of Christians in Africa has been gradually decreasing over the years,
#starting at 77% in 2009 and reaching 65% in 2018. 
#the percentage of Unaffiliated individuals has been slowly increasing, going from 17% in 2009 to 26% in 2018 and this is wen
#using fictional data
```
